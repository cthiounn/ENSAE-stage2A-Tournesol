```python
import pandas as pd
```

```python
df_scaling= pd.read_csv("scaling_after")
df_scaling=df_scaling.drop(columns=['id','scale_uncertainty','translation_uncertainty','poll_id','user_id'])
df_scaling=df_scaling.set_index('criteria')
df_scaling
```

```python
df_ratings_after= pd.read_csv("ratings_after")
df_ratings_after=df_ratings_after.drop(columns=['id','contributor_rating_id'])
df_ratings_after=df_ratings_after.set_index("criteria")
df_ratings_before= pd.read_csv("ratings_before")
df_ratings_before=df_ratings_before.drop(columns=['id','contributor_rating_id'])
df_ratings_before=df_ratings_before.set_index("criteria")
```

```python
mega_df=df_ratings_after.join(df_ratings_before, how='inner',lsuffix='_a',rsuffix='_b')
mega_df_with_scale=mega_df.join(df_scaling,how='inner')
mega_df_with_scale
```

```python
df=mega_df_with_scale
df["compute_scaled_score"] = df["scale"] * df["raw_score_a"] + df["translation"]
df["compute_scaled_uncertainty"] = df['raw_uncertainty_a']*df["scale"]
df
```
